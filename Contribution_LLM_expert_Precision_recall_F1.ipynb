## Domain contribution LLM evaluations metrics

```python
files_LLM = {
    "Metagenomic": "Metagenomic_confusion_metrics_per_tool_annotation.tsv",
    "Phylogeny": "Phylogeny_confusion_metrics_per_tool_annotation.tsv",
    "Single_Cell": "Single_Cell_confusion_metrics_per_tool_annotation.tsv",
    "Systems_biology": "Systems_biology_confusion_metrics_per_tool_annotation.tsv",
    "Neuroimaging": "Neuroimaging_confusion_metrics_per_tool_annotation.tsv",
    "Genetic_variant": "Genetic_variant_confusion_metrics_per_tool_annotation.tsv"
}
```

```python
def summarize_metrics(tsv_path, domain_name):
    df = pd.read_csv(tsv_path, sep="\t")

    return {
        "domain": domain_name,
        "mean_precision": df["precision"].mean(),
        "std_precision": df["precision"].std(),
        "mean_recall": df["recall"].mean(),
        "std_recall": df["recall"].std(),
        "mean_f1": df["f1_score"].mean(),
        "std_f1": df["f1_score"].std(),
    }

# compute summaries
summaries = [
    summarize_metrics(path, domain)
    for domain, path in files_LLM.items()
]

summary_df_LLM = pd.DataFrame(summaries).round(3)
```

```python
def plot_all_domains_all_metrics_with_values(df):
    domains = [d for d in df["domain"].unique() if d != "GLOBAL"] + ["GLOBAL"]
    metrics = ["Precision", "Recall", "F1"]

    x_labels = []
    for d in domains:
        for m in metrics:
            x_labels.append(f"{d}\n{m}")

    x = np.arange(len(x_labels))
    width = 0.35

    llm_vals, llm_errs = [], []
    expert_vals, expert_errs = [], []

    for d in domains:
        for m in metrics:
            llm = df[(df.domain == d) & (df.metric == m) & (df.model == "LLM")]
            expert = df[(df.domain == d) & (df.metric == m) & (df.model == "Expert")]

            llm_vals.append(llm["mean"].values[0])
            llm_errs.append(llm["std"].values[0])
            expert_vals.append(expert["mean"].values[0])
            expert_errs.append(expert["std"].values[0])

    plt.figure(figsize=(18, 7))

    bars_llm = plt.bar(
        x - width / 2,
        llm_vals,
        width,
        yerr=llm_errs,
        capsize=4,
        label="LLM",
    )

    bars_expert = plt.bar(
        x + width / 2,
        expert_vals,
        width,
        yerr=expert_errs,
        capsize=4,
        label="Expert",
    )

    # ---- annotate values under bars ----
    def annotate_bars(bars, means, stds):
        for bar, mean, std in zip(bars, means, stds):
            x_pos = bar.get_x() + bar.get_width() / 2
            plt.text(
                x_pos,
                -0.08,                       # below x-axis
                f"{mean:.2f}\n± {std:.2f}",
                ha="center",
                va="top",
                fontsize=8,
            )

    annotate_bars(bars_llm, llm_vals, llm_errs)
    annotate_bars(bars_expert, expert_vals, expert_errs)

    plt.xticks(x, x_labels, rotation=45, ha="right")
    plt.ylabel("Score")
    plt.ylim(-0.15, 1)  # space for text underneath
    plt.title("Precision / Recall / F1 by Domain — LLM vs Expert (with Global)")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_metric_by_domain_with_values(df, metric_name):
    """
    df must contain:
    columns = [domain, metric, model, mean, std]
    metric_name = "Precision", "Recall", or "F1"
    """

    domains = [d for d in df["domain"].unique() if d != "GLOBAL"] + ["GLOBAL"]
    x = np.arange(len(domains))
    width = 0.35

    llm_vals, llm_errs = [], []
    expert_vals, expert_errs = [], []

    for d in domains:
        llm = df[(df.domain == d) & 
                 (df.metric == metric_name) & 
                 (df.model == "LLM")]
        
        expert = df[(df.domain == d) & 
                    (df.metric == metric_name) & 
                    (df.model == "Expert")]

        llm_vals.append(llm["mean"].values[0])
        llm_errs.append(llm["std"].values[0])
        expert_vals.append(expert["mean"].values[0])
        expert_errs.append(expert["std"].values[0])

    plt.figure(figsize=(14, 6))

    bars_llm = plt.bar(
        x - width/2,
        llm_vals,
        width,
        yerr=llm_errs,
        capsize=4,
        label="LLM"
    )

    bars_expert = plt.bar(
        x + width/2,
        expert_vals,
        width,
        yerr=expert_errs,
        capsize=4,
        label="Expert"
    )

    # ---- annotate values under bars ----
    def annotate_bars(bars, means, stds):
        for bar, mean, std in zip(bars, means, stds):
            x_pos = bar.get_x() + bar.get_width()/2
            plt.text(
                x_pos,
                -0.08,
                f"{mean:.2f}\n± {std:.2f}",
                ha="center",
                va="top",
                fontsize=9
            )

    annotate_bars(bars_llm, llm_vals, llm_errs)
    annotate_bars(bars_expert, expert_vals, expert_errs)

    plt.xticks(x, domains, rotation=45, ha="right")
    plt.ylabel("Score")
    plt.ylim(-0.15, 1)
    plt.title(f"{metric_name} by Domain — LLM vs Expert")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def summarize_per_annotation_and_global(tsv_path, domain_name):
    df = pd.read_csv(tsv_path, sep="\t")
    
    # --- per annotation_type ---
    per_ann_list = []
    for ann in df["annotation_type"].unique():
        sub = df[df["annotation_type"]==ann]
        per_ann_list.append({
            "domain": domain_name,
            "annotation_type": ann,
            "type": "annotation_type",
            "mean_precision": sub["precision"].mean(),
            "std_precision": sub["precision"].std(),
            "mean_recall": sub["recall"].mean(),
            "std_recall": sub["recall"].std(),
            "mean_f1": sub["f1_score"].mean(),
            "std_f1": sub["f1_score"].std(),
        })
    per_ann = pd.DataFrame(per_ann_list)
    
    # --- global per domain ---
    global_row = pd.DataFrame([{
        "domain": domain_name,
        "annotation_type": "GLOBAL",
        "type": "global",
        "mean_precision": df["precision"].mean(),
        "std_precision": df["precision"].std(),
        "mean_recall": df["recall"].mean(),
        "std_recall": df["recall"].std(),
        "mean_f1": df["f1_score"].mean(),
        "std_f1": df["f1_score"].std(),
    }])
    
    # combine
    combined = pd.concat([per_ann, global_row], ignore_index=True)
    return combined
```

```python
all_summaries = []
for domain, path in files.items():
    all_summaries.append(summarize_per_annotation_and_global(path, domain))

summary_df_LLM = pd.concat(all_summaries, ignore_index=True)
summary_df_LLM = summary_df_LLM.round(3)
summary_df_LLM
```

```python
def plot_all_metrics_long(df):
    # Create combined x-axis: domain + annotation_type
    x_labels = []
    domains = df["domain"].unique()
    annotation_types = df["annotation_type"].unique()
    metrics = ["Precision","Recall","F1"]

    # order: all annotation types + GLOBAL per domain
    x_pos_labels = []
    for domain in domains:
        annots = df[df.domain == domain]["annotation_type"].unique()
        annots = [a for a in annots if a != "GLOBAL"] + ["GLOBAL"]
        for a in annots:
            for m in metrics:
                x_pos_labels.append(f"{domain}\n{a}\n{m}")

    x = np.arange(len(x_pos_labels))
    width = 0.35

    # fill bar values in order
    llm_vals = []
    llm_errs = []
    expert_vals = []
    expert_errs = []

    for domain in domains:
        annots = df[df.domain == domain]["annotation_type"].unique()
        annots = [a for a in annots if a != "GLOBAL"] + ["GLOBAL"]
        for a in annots:
            for m in metrics:
                llm = df[(df.domain==domain)&(df.annotation_type==a)&(df.metric==m)&(df.model=="LLM")]
                expert = df[(df.domain==domain)&(df.annotation_type==a)&(df.metric==m)&(df.model=="Expert")]
                llm_vals.append(llm["mean"].values[0])
                llm_errs.append(llm["std"].values[0])
                expert_vals.append(expert["mean"].values[0])
                expert_errs.append(expert["std"].values[0])

    plt.figure(figsize=(20,7))
    bars_llm = plt.bar(x - width/2, llm_vals, width, yerr=llm_errs, capsize=3, label="LLM")
    bars_expert = plt.bar(x + width/2, expert_vals, width, yerr=expert_errs, capsize=3, label="Expert")

    # annotate mean ± std under bars
    def annotate_bars(bars, means, stds):
        for bar, mean, std in zip(bars, means, stds):
            x_pos = bar.get_x() + bar.get_width()/2
            plt.text(x_pos, -0.08, f"{mean:.2f}\n±{std:.2f}", ha="center", va="top", fontsize=7)

    annotate_bars(bars_llm, llm_vals, llm_errs)
    annotate_bars(bars_expert, expert_vals, expert_errs)

    plt.xticks(x, x_pos_labels, rotation=60, ha="right", fontsize=8)
    plt.ylabel("Score")
    plt.ylim(-0.15,1)
    plt.title("Precision / Recall / F1 by Domain and Annotation Type — LLM vs Expert")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plot_domain_annotation(df, domain):
    """
    Creates a separate figure for each domain, showing Precision/Recall/F1 for each annotation_type.
    """
    df_dom = df[df.domain == domain]
    annots = df_dom[df_dom.type=="annotation_type"]["annotation_type"].unique()
    metrics = ["Precision","Recall","F1"]
    
    for a in annots:
        df_plot = df_dom[(df_dom.annotation_type==a)]
        x = np.arange(len(metrics))
        width = 0.35
        
        llm_vals = [df_plot[(df_plot.model=="LLM") & (df_plot.metric==m)]["mean"].values[0] for m in metrics]
        llm_errs = [df_plot[(df_plot.model=="LLM") & (df_plot.metric==m)]["std"].values[0] for m in metrics]
        
        expert_vals = [df_plot[(df_plot.model=="Expert") & (df_plot.metric==m)]["mean"].values[0] for m in metrics]
        expert_errs = [df_plot[(df_plot.model=="Expert") & (df_plot.metric==m)]["std"].values[0] for m in metrics]
        
        plt.figure(figsize=(6,4))
        bars_llm = plt.bar(x - width/2, llm_vals, width, yerr=llm_errs, capsize=4, label="LLM")
        bars_expert = plt.bar(x + width/2, expert_vals, width, yerr=expert_errs, capsize=4, label="Expert")
        
        # annotate mean ± std under bars
        for bar, mean, std in zip(list(bars_llm)+list(bars_expert), llm_vals+expert_vals, llm_errs+expert_errs):
            x_pos = bar.get_x() + bar.get_width()/2
            plt.text(x_pos, -0.08, f"{mean:.2f}\n±{std:.2f}", ha="center", va="top", fontsize=8)
        
        plt.xticks(x, metrics)
        plt.ylabel("Score")
        plt.ylim(-0.15,1)
        plt.title(f"{domain} — {a}")
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
for domain in long_df["domain"].unique():
    plot_domain_annotation(long_df, domain)
```
