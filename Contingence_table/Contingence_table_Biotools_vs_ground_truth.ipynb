```python

# Load TSV
df = pd.read_csv("EDAM_terms_URI_ground_truth_validated.tsv", sep="\t")

# Extract unique URIs
gt_uri_list = df["EDAM term URI"].dropna().unique().tolist()

print(len(gt_uri_list))
```

```python
uri_data_format = set()

with open("List_data_format_tools.txt") as f:
    for line in f:
        line = line.strip()
        if line.startswith("http"):
            uri_data_format.add(line)

uri_data_format_bio_tools = list(uri_data_format)

print(len(uri_data_format_bio_tools))
```

```python

with open("Direct_EDAM_annotation_all_tools_bio_tools.txt") as f:
    data = ast.literal_eval(f.read())

uri_biotools = set()

for tool in data.values():
    for category in tool.values():
        for entry in category:
            uri_biotools.add(entry["URI"])

uri_topic_operation_biotools = list(uri_biotools)

print(len(uri_topic_operation_biotools))
```

```python
all_uri_biotools_tools = list(set(uri_topic_operation_biotools) | set(uri_data_format_bio_tools))

print(len(all_uri_biotools_tools))
```

```python
def build_edam_contingency_table(gt_list, biotools_list, neighbors_gt, neighbors_biotools):
    """
    Build a contingency table comparing GT and BioTools EDAM annotations
    with and without neighbors.
    """

    gt_set = set(gt_list)
    biotools_set = set(biotools_list)

    neighbors_gt_set = set(neighbors_gt)
    neighbors_biotools_set = set(neighbors_biotools)

    # Direct annotations
    only_gt = len(gt_set - biotools_set)
    common_direct = len(gt_set & biotools_set)
    only_biotools = len(biotools_set - gt_set)

    # Direct + neighbors
    only_gt_neighbors = len(neighbors_gt_set - neighbors_biotools_set)
    common_neighbors = len(neighbors_gt_set & neighbors_biotools_set)
    only_biotools_neighbors = len(neighbors_biotools_set - neighbors_gt_set)
    
    only_gt = len(gt_set - biotools_set)
    only_gt_neighbors = len(neighbors_gt_set - neighbors_biotools_set)
    only_gt_difference = only_gt - only_gt_neighbors
    print(only_gt_difference)

    table = pd.DataFrame(
        [
            [only_gt, common_direct, only_biotools],
            [only_gt_neighbors, common_neighbors, only_biotools_neighbors]
        ],
        index=["Direct annotations", "Direct annotations + neighbors"],
        columns=[
            "Only GT",
            "Common (GT ∩ BioTools)",
            "Only BioTools"
        ]
    )

    return table
```

```python
table = build_edam_contingency_table(
    gt_uri_list,
    all_uri_biotools_tools,
    neighbors_gt,
    neighbors_biotools
)

table
```
